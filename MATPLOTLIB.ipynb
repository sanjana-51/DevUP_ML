Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

```python
x=np.array([0,6])
y=np.array([0,250])
```

```python
plt.plot(x,y)
plt.show()
```

```python
xpoint=np.array([1,2,8,9,12])
ypoint=np.array([3,8,5,18,50])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
plt.plot(xpoint,ypoint,linestyle='dotted',color='red')
plt.show()
```

```python
plt.plot(xpoint,ypoint,color='red',linewidth='20')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
y3=np.array([1,2,3,4])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 180, 210, 250, 280, 320,500 ])
plt.plot(x,y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.savefig('Graph.png')
plt.show()
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title("GRAPH")
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.plot(y_sin,y_cos)
```

```python
## BAR PLOT
x1 = [2,8,10,12]
y1=[11,16,8,4]
plt.bar(x1,y1)
```

```python
# HISTOGRAM 
a=np.array([23,45,6,65,5,43,33,54,98,12])
plt.hist(a,bins=20)
plt.show()
```

```python
# Data to plot
labels=['Python','C++','Ruby','Java']
sizes=[215,150,123,231]
colors=['red','yellow','blue','green']
plt.pie(sizes,labels=labels,colors=colors,shadow=True)
plt.show()
```
